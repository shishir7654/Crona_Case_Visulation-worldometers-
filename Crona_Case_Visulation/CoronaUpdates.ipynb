```python
df['NewDeaths'] = df['NewDeaths'].apply(lambda s:"".join(s.split(',')) if type(s) == str else s).astype(float)
df = df[~df['NewDeaths'].isna()]

df.sort_values('NewDeaths', ascending=False, inplace=True)
df = df[['Country', 'NewDeaths']]
df.sort_values('NewDeaths', ascending=False, inplace=True)
```
